```python
# -*- coding: utf-8 -*-
"""Instructor Demo: CSV Reader.

This script will use the Pathlib library to set the file path,
use the csv library to read in the file, iterate over each
row of the file to capture P/L dates, calculate avg rate of change,
greatest & smalled increase and write data to file.
to a csv file.
"""

# Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path
csvpath = Path('Bank/Resources/budget_data.csv')

# Initialize P/L variable
##prev_mypls = []
mypls = []

# Initialize line_num variable
line_num = 0
count = 0

# Open the input path as a file object
with open(csvpath, 'r') as csvfile:

    # Print the datatype of the file object
    print(type(csvfile))

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    # Print the datatype of the csvreader
    print(type(csvreader))

    # Go to the next row from the start of the file
    # (which is often the first row/header) and iterate line_num by 1
    header = next(csvreader)
    line_num += 1
    # Print the header
    #print(f"{header} <---- HEADER")

    # Read each row of data after the header
    for row in csvreader:
        # Print the row
        #print(row)
        # Set date variable equal to the value in the 1st column of each row
        mydate = str(row[0])
        mypl = int(row[1])
        # Set date variable equal to the value in the previous 1st column of each row
        ##prev_mypl = int(row[1])
        # Append the row PL values to the list of PLs
        mypls.append(mypl)
        mydates.append(mydate)
        ##prev_mypls.append(prev_mypl)
        # Cumulatively sum up the total and count of PLs
        count += 1

# Print out the summary statistics
print("---------Summary Statistics----------")
print(f"Number of Months: {count}")

# Initialize metric variables
acr_mypl = 0
totalcr_mypl = 0
cr_mypl = 0
curr_mypl = 0
prev_mypl = 0
count_mypl = 0
max_mypl = 0
min_mypl = 0


# Calculate the max, min, and average of the list of PLs
for mypl in mypls:

    # Sum the count variables
    count_mypl += 1
    
    # Logic to determine period-to-period rate of change
    if curr_mypl == 0:
        curr_mypl = mypl
    elif curr_mypl != mypl:
        prev_mypl = curr_mypl
        curr_mypl = mypl
        
        # Sum the total of change rates
        cr_mypl = (curr_mypl - prev_mypl)
        totalcr_mypl += cr_mypl
        #print(f"curr: {curr_mypl}")
        #print(f"prev: {prev_mypl}")
        #print(f"totcr: {totalcr_mypl}")
        #print(f"totcr: {cr_mypl}")
        #print(f"my_date: {mydate}")

    # Logic to determine max PL decrease & increase
    if min_mypl == 0 and cr_mypl < 0:
        min_mypl = cr_mypl
    elif cr_mypl < 0 and cr_mypl < min_mypl:
        min_mypl = cr_mypl
        min_date = mydates[count_mypl-1]
        #print(f"mini_date: {mydates[count_mypl-1]}")
    elif max_mypl == 0 and cr_mypl > 0:
        max_mypl = cr_mypl
    elif cr_mypl > 0 and cr_mypl > max_mypl:
        max_mypl = cr_mypl
        max_date = mydates[count_mypl-1]
        #print(f"maxi_date: {mydates[count_mypl-1]}")

# Calculate the average PL, round to the nearest 2 decimal places
acr_mypl = round(totalcr_mypl / (count_mypl-1), 2)

# Print the metrics
print(f"Total Months: ${total_mypl}")
print(f"Average Change: ${acr_mypl}")
print(f"Greatest Increase in Profits:{max_date} (${max_mypl})")
print(f"Greatest Decrease in Profits:{min_date} (${min_mypl})")
#print(max_mypl, min_mypl, avg_mypl)
#print(f"Total PL Avgs: {totalcr_mypl}")

line1 = (f"---------Summary Statistics----------\n")
line2 = (f"Number of Months: {count}\n")
line3 = (f"Total Months: ${total_mypl}\n")
line4 = (f"Average Change: ${acr_mypl}\n")
line5 = (f"Greatest Increase in Profits:{max_date} (${max_mypl})\n")
line6 = (f"Greatest Decrease in Profits:{min_date} (${min_mypl})\n")

# Set the output file path
output_path = Path('PyBnkoutput.txt')

# Open the output path as a file object
with open(output_path, 'w') as out:
    # Write to file object
   #file.write(f"Total Months: ${total_mypl}\n")
    out.writelines([line1, line2, line3, line4, line5, line6])
```
